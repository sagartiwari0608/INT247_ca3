```python
import pandas as pd
import numpy as np

df=pd.read_csv("F:\mlProject\letter-recognition.csv")
df
```

```python
df.info()
```

```python
letters= list(np.sort(df['letter'].unique()))
print(letters)
```

```python
df_mean= df.groupby("letter").mean()
df_mean
```

```python
x=df.drop(["letter"],axis=1)
y=df['letter']
print("this is data X for train test split ", x)
print("this is data Y for test ", y)

```

```python
from sklearn.preprocessing import scale
x=scale(x)
print(x)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test, y_train, y_test =train_test_split(x,y,train_size=0.7,test_size=0.3, random_state=150)
```

```python
from sklearn.svm import SVC 

# Building a linear SVM model
model= SVC()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix

# accuracy and confusion matrix
print("accuracy for SVM :", metrics.accuracy_score(y_true = y_test,y_pred = y_pred), "\n")
print("confusion_matrix for SVM :", metrics.confusion_matrix(y_true = y_test,y_pred = y_pred))
```
